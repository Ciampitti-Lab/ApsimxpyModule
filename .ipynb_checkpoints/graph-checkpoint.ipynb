```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_parquet("/workspace/merged_results.parquet",engine="fastparquet")
df.columns

df["Clock.Today"] = pd.to_datetime(df["Clock.Today"])


pivot_df = df.pivot(index="Clock.Today",
                    columns="n_sim",
                    values="SurfaceOrganicMatter.Cover")

plt.figure(figsize=(12,6))
for col in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[col], label=f"Sim {col}")

plt.xlabel("Date")
plt.ylabel("Variable")
plt.legend(title="Simulation")
plt.tight_layout()
plt.show()

```
